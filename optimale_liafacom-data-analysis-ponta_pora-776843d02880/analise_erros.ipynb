```python
import pandas as pd 
import numpy as np
import pandas
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# final result
n_groups = 3

vazao = (10.331901903725626, 30.204604246386349,  1.5550022323135584)
pressao_entrada = (2.4574397940685433, 1.5024222142450667, 1.8023970223167265)
pressao_media = (4.9745525312357284, 2.7053020718338385, 2.1539635724663913)
pressao_alta = (7.879857939249737, 11.93156304093346, 3.3103726169878556)

#means_women = (25, 32, 34, 20, 25)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.5

opacity = 0.5
error_config = {'ecolor': '0.3'}
fig.set_size_inches(15.5, 10.5)
rects1 = plt.bar(index, vazao, (bar_width/2)-0.1,
                 alpha=opacity,
                 color='b',
                 error_kw=error_config,
                 label='Vazao')

rects2 = plt.bar(index + bar_width/2 - 0.1, pressao_entrada, (bar_width/2)-0.1,
                 alpha=opacity,
                 color='r',
                 error_kw=error_config,
                 label='Pressao Entrada')
rects3 = plt.bar(index + bar_width/2 + bar_width/2 - 0.2, pressao_media, (bar_width/2)-0.1,
                 alpha=opacity,
                 color='g',
                 error_kw=error_config,
                 label='Pressao Media')
rects4 = plt.bar(index + bar_width/2 + bar_width/2 + bar_width/2 - 0.3, pressao_alta, (bar_width/2)-0.1,
                 alpha=opacity,
                 color='black',
                 error_kw=error_config,
                 label='Pressao Alta')


plt.xlabel('Models')
plt.ylabel('Scores')
plt.ylim([0, 35])
plt.title('Errors for 3 methods (Janelas deslizantes)')
plt.xticks(index + bar_width/1.6, ('KNN', 'SVM','LSTM'))

plt.legend()
plt.plot()
```
