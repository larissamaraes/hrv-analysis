```python
import pandas as pd 
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('690g-pressao_entrada-vila-aurea-20-04-2016.csv')

data_matrix = data.as_matrix()

data_matrix = data_matrix[331982:,2:]        # Elimina os termos nan da matriz

#data_matrix = np.nan_to_num(data_matrix)    # Tentei usar este comando para eliminar o 
                                             # termo nan, porém não funcionou                              
minutos = data_matrix.shape[0]

print "Total de minutos:", minutos

print data_matrix
```

```python
# Redimensiona a matriz de acordo com a janela de m minutos,
# para isso acrescenta zeros nas últimas linhas, se necessário; 

def reshape_matrix (data_matrix, minutos, m):

    resto = minutos % m
    num_zeros = m - resto
    
    if(num_zeros != 0):
        zeros = np.zeros((num_zeros,1))
        data_min = np.concatenate((data_matrix, zeros), axis = 0)
    
    data_min = data_min.reshape(-1, m)
    
    return data_min
```

# De 1 em 1 minuto

```python
x = np.arange(1, minutos + 1, 1)
y = data_matrix

fig = plt.gcf()
    
plt.plot(x, y, color = 'blue')

plt.title(u'Pressão a cada minuto com desvio padrão', fontsize = 20)
plt.xlabel('Minutos', fontsize = 20)
plt.ylabel(u'Pressão',  fontsize = 20)
fig.set_size_inches(18.5, 14.5)

plt.show()
```

```python
# De 1 em 1 minuto com zoom

porcentagem = 1
i = int(minutos * 0.01 * porcentagem)
indice = minutos - i

x = np.arange(1, i + 1, 1)
y = data_matrix[indice:,:]

fig = plt.gcf()
    
plt.plot(x, y, color = 'blue')

plt.title(u'Pressão a cada minuto com desvio padrão (zoom)', fontsize = 20)
plt.xlabel('Minutos', fontsize = 20)
plt.ylabel(u'Pressão',  fontsize = 20)
fig.set_size_inches(18.5, 14.5)

plt.show()
```

# De 2 em 2 minutos

```python
data_2min = reshape_matrix(data_matrix, minutos, 2)

#print (data_2min)

mean_2min = data_2min.mean(axis=1)
std_2min = data_2min.std(axis=1, dtype=float)

n = mean_2min.shape[0]

fig = plt.gcf()
    
x = np.arange(1, n + 1, 1) 
y = mean_2min

plt.errorbar(x, y, yerr = std_2min, color = 'red')
   
plt.title(u'Média da pressão de 2 em 2 minutos com desvio padrão', fontsize = 20)
plt.xlabel('Minutos', fontsize = 20)
plt.ylabel(u'Pressão',  fontsize = 20)
    
fig.set_size_inches(18.5, 14.5)
plt.show()
    
```

```python
# De 2 em 2 minutos com zoom 

data_2min = reshape_matrix(data_matrix, minutos, 2)

mean_2min = data_2min.mean(axis=1)
std_2min = data_2min.std(axis=1, dtype=float)

n = mean_2min.shape[0]

porcentagem = 1
i = int(n * 0.01 * porcentagem)
indice = n - i

x = np.arange(1, i + 1, 1)
y = mean_2min[indice:]    

fig = plt.gcf()

plt.errorbar(x, y, yerr = std_2min[indice:], color = 'red')
   
plt.title(u'Média da pressão de 2 em 2 minutos com desvio padrão (zoom)', fontsize = 20)
plt.xlabel('Minutos', fontsize = 20)
plt.ylabel(u'Pressão',  fontsize = 20)
   
fig.set_size_inches(18.5, 14.5)
plt.show()
```
