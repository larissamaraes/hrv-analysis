```python
"""

Author: Optimale
Date:   26-04-17

"""
%matplotlib inline
import pandas
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsRegressor
import time
from sklearn.svm import SVR
from sklearn.model_selection import RandomizedSearchCV
import datetime
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.arima_model import ARIMA
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from estatisticas_time_serie import create_dataset, general_forecasting, smooth
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf
import warnings
warnings.filterwarnings('ignore')

from sklearn.metrics import explained_variance_score, r2_score, mean_absolute_error, mean_squared_error

# load the first dataset vilacarvalho-vazao-60
dataframe = pandas.read_csv('690a-vazao-vila-aurea-20-04-2016.csv', engine='python', skipfooter=3)
data_matrix = dataframe.as_matrix()
data_matrix = data_matrix[340519:,2:]

data_convolution = np.array(data_matrix).ravel()

timeValues = dataframe['Time']
values     = dataframe['Value']

timestamps=[]
newV=[]
for idx,val in enumerate(data_convolution):
    if not data_convolution[idx] == 'null' and not np.isnan(data_convolution[idx]): 
        newV.append(float(data_convolution[idx]))
        value = time.mktime(datetime.datetime.strptime(timeValues[idx], "%Y-%m-%d %H:%M:%S").timetuple())
        timestamps.append([value])

newV = np.array(newV)
def reject_outliers(data, m=3):
    return data[abs(data - np.mean(data)) < m * np.std(data)]
newV = reject_outliers(newV)

plt.figure(num=None, figsize=(12, 4), dpi=80, facecolor='w', edgecolor='k')
plt.plot(newV,label='values')
plt.title('Time Serie of Vila Aurea Normalized')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
# preprocessing
data_convolution1 = smooth(newV, 10, window='hanning')



plt.figure(num=None, figsize=(23, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(data_convolution1,label='values')
plt.title('Time Serie after smoothing',fontsize=24)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
# splitting data
train_size = int(len(data_convolution1) * 0.8)
test_size = len(data_convolution1) - train_size
train, test = data_convolution1[0:train_size], data_convolution1[train_size:train_size+test_size]
print(len(train), len(test))
```

```python
# create dataset 
# Using  KNN
look_back = 420
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)

# Ssing lstm
scaler = MinMaxScaler(feature_range=(0, 1))
# normalize the dataset
datasetLstmTrain = scaler.fit_transform(train)
datasetLstmTest = scaler.fit_transform(test)
look_back = 420
trainX_lstm, trainY_lstm = create_dataset(datasetLstmTrain, look_back)
testX_lstm, testY_lstm = create_dataset(datasetLstmTest, look_back)
# reshape input to be [samples, time steps, features]
trainX_lstm = np.reshape(trainX_lstm, (trainX_lstm.shape[0], 1, trainX_lstm.shape[1]))
testX_lstm = np.reshape(testX_lstm, (testX_lstm.shape[0], 1, testX_lstm.shape[1]))

# using ARIMA
#s = pd.Series(train, index=timeValues[0:train_size])
```

```python
#train the models
# KNN
clf_KNN = KNeighborsRegressor(n_neighbors=1)
clf_KNN.fit(trainX, trainY)

# SVM

clf_SVM = SVR(verbose=True)
#params_list = {"C": [1e0, 1e1, 1e2, 1e3],"gamma": np.logspace(-2, 2, 5)}
#n_iter_search=10
#random_search = RandomizedSearchCV(clf_SVM, param_distributions=params_list,
#                                   n_iter=5)
#random_search.fit(trainX, trainY)
#clf_SVM = random_search.best_estimator_
print("training svm")
clf_SVM.fit(trainX,trainY)

print("training lstm")
#LSTM
model = Sequential()
model.add(LSTM(10, input_dim=look_back))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainX_lstm, trainY_lstm, nb_epoch=10, batch_size=128, verbose=1)
# Arima
# fitting the model
```

```python
#predicting
forecast_KNN = general_forecasting(testX, clf_KNN, size_forecasting=420 ,method='knn', strategy_approach='retro')
forecast_LSTM = general_forecasting(testX_lstm, model, size_forecasting=420,method='lstm', strategy_approach='retro')
forecast_SVM =  general_forecasting(testX, clf_SVM, size_forecasting=420, strategy_approach='retro')
#forecastedArima = general_forecasting(test,train=s.values,size_forecasting=420 ,classifier=None, method='arima', strategy_approach='retro')
```

```python
# rescaling back to initial dimension
forecast_LSTM = scaler.inverse_transform(forecast_LSTM)
testY_lstm = scaler.inverse_transform(testY_lstm)
```

```python
#scores
print("mean_squared_error  KNN = ",mean_squared_error(testY[:420],forecast_KNN))
print("mean_squared_error  SVM = ",mean_squared_error(testY[:420],forecast_SVM))
print("mean_squared_error  LSTM = ",mean_squared_error(testY_lstm[:420],forecast_LSTM))
#print("mean_squared_error  ARIMA = ",mean_squared_error(test[:48],forecastedArima))
```
