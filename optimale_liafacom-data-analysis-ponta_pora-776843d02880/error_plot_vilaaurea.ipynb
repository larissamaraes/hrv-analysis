```python
import pandas as pd 
import numpy as np
import pandas
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# final result
n_groups = 3

vazao = (17.891527823969316, 9.0410888937649805, 138.59284512113416)
pressao = (1.9966839079889538, 0.8079555164278972, 24.363521582808598)
pressao_media = (2.4399868984345003, 0.88205661154327086, 24.365338221704022)
pressao_alta = (3.5722515964259154, 1.3102659124277587, 1087.8469509053964)

#means_women = (25, 32, 34, 20, 25)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.5

opacity = 0.5
error_config = {'ecolor': '0.3'}
fig.set_size_inches(15.5, 10.5)
rects1 = plt.bar(index, vazao, (bar_width/2)-0.1,
                 alpha=opacity,
                 color='b',
                 error_kw=error_config,
                 label='Vazao')

rects2 = plt.bar(index + bar_width/2 - 0.1, pressao, (bar_width/2)-0.1,
                 alpha=opacity,
                 color='r',
                 error_kw=error_config,
                 label='Pressao')
rects3 = plt.bar(index + bar_width/2 + bar_width/2 - 0.2, pressao_media, (bar_width/2)-0.1,
                 alpha=opacity,
                 color='g',
                 error_kw=error_config,
                 label='Pressao Media')
rects4 = plt.bar(index + bar_width/2 + bar_width/2 + bar_width/2 - 0.3, pressao_alta, (bar_width/2)-0.1,
                 alpha=opacity,
                 color='black',
                 error_kw=error_config,
                 label='Pressao Alta')


plt.xlabel('Models')
plt.ylabel('Scores')
plt.ylim([0, 100])
plt.title('Errors for 3 methods (Janelas deslizantes)')
plt.xticks(index + bar_width/1.6, ('KNN', 'LSTM', 'ARIMA'))

plt.legend()
plt.plot()
```
